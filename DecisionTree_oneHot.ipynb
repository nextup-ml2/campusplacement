```python
def one_hot(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode 
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
d2=one_hot(d1,['gender', 'pu_stream', 'deg_stream', 'workex', 'status'])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X = d2[['tenth_marks', 'pu_marks', 'deg_marks', 'gender_F', 'gender_M', 'pu_stream_Arts', 'pu_stream_Commerce', 'pu_stream_Science', 'deg_stream_Comm&Mgmt', 'deg_stream_Others', 'deg_stream_Sci&Tech','workex_No', 'workex_Yes']]
y = d2[['status_Not Placed', 'status_Placed']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtr = DecisionTreeClassifier()
dtr.fit(X_train,y_train)
```

```python
y_pred = dtr.predict(X_train)
accuracy_score(y_train, y_pred)
```

```python
y_pred = dtr.predict(X_test)
accuracy_score(y_test, y_pred)
```
